```python
# Univariate Analysis for Numeric Variables using Boxplot

# Mengambil fitur numerik
numeric_features = df.select_dtypes(include=[np.number])

# Menentukan ukuran grid untuk subplot secara dinamis
num_features = len(numeric_features.columns)
cols = 3
rows = math.ceil(num_features / cols)

plt.figure(figsize=(cols * 4, rows * 4))
for i, column in enumerate(numeric_features.columns, 1):
    plt.subplot(rows, cols, i)
    sns.boxplot(y=column, data=df, color='#019fab')
    plt.title(f'Boxplot {column}')
plt.tight_layout()
plt.show()
```

- Age dan Daily Time Spent on Site juga memiliki korelasi negatif (-0.33). Ini menunjukkan tren yang serupa, yaitu semakin tua seseorang, semakin sedikit waktu yang mereka habiskan di situs tertentu.

- Area Income memiliki korelasi yang relatif lemah dengan fitur lainnya. Ini menunjukkan bahwa pendapatan tidak memiliki hubungan yang kuat dengan fitur-fitur numerical lainnya.

# 2. Data Cleaning & Preprocessing

## 2.1 Data Cleaning

```python
df_clean = df.copy()
```

### 2.1.1 Handling Missing Value

```python
result = df.loc[df['Area Income'].isna()]
result[['Daily Time Spent on Site', 'Age', 'Area Income']][:7]
# result
```

```python
# Pengecekan data yang missing value
print("\nMissing values:")
print(df.isnull().sum())
```

```python
df_clean['Area Income'].mean()
```

```python
# Mengisi nilai null pada Kolom 'Daily Time Spent on Site' dengan Mean
df_clean['Daily Time Spent on Site'].fillna(df_clean['Daily Time Spent on Site'].mean(), inplace=True)

# Mengisi nilai null pada Kolom 'Area Income' dengan Mean
df_clean['Area Income'].fillna(df_clean['Area Income'].mean(), inplace=True)

# Mengisi nilai null pada Kolom 'Daily Internet Usage' dengan Mean
df_clean['Daily Internet Usage'].fillna(df_clean['Daily Internet Usage'].mean(), inplace=True)

# Menghapus data yang nilai null pada Kolom 'Male'
df_clean.dropna(subset=['Male'], inplace=True)
```

- Melakukan handling missing value pada fitur Daily Time Spent on Site, Area Income, Daily Internet Usage dengan mengisinya menggunakan mean.

- Melakukan handling missing value pada fitur Male dengan menghapus data missing value.

## 2.2 Data Preprocessing

### 2.2.1 Drop Features

```python
# Menghapus Fitur yang Tidak Diperlukan
drop_features = ['Unnamed: 0']

# df.drop(drop_features, axis=1, inplace=True)
df_clean = df_clean.drop(drop_features, axis=1, inplace=False)
df_clean
```

### 2.2.2 Feature Encoding

```python
# Tampilkan kolom dengan tipe data object (kategorikal)
print("\nCategorical Features:", categorical_features)
```

```python
# Menggunakan get_dummies untuk melakukan one-hot encoding pada data categorical
df_clean = pd.get_dummies(df_clean, columns=['Male'], prefix=['Gender'])
df_clean = pd.get_dummies(df_clean, columns=categorical_features[1:])
```

```python
print(f'Result one-hote encoding:')
print(df_clean.columns)
```

### 2.2.4 Extraction Datetime

```python
# Mengubah tipe data pada fitur Timestamp menjadi datetime
df_clean['Timestamp'] = pd.to_datetime(df_clean['Timestamp'])
```

```python
# Membuat kolom baru berdasarkan data waktu
df_clean['Year'] = df_clean['Timestamp'].dt.year
df_clean['Month'] = df_clean['Timestamp'].dt.month
df_clean['Week'] = df_clean['Timestamp'].dt.isocalendar().week
df_clean['Day'] = df_clean['Timestamp'].dt.day
```
